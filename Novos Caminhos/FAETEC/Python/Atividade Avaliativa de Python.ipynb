1) Faça um aplicativo em Python que leia a idade de uma pessoa expressa em anos, meses e dias e mostre-a expressa em dias. Leve em consideração o ano com 365 dias e o mês com 30.

```python
anos = int(input("Digite sua idade em Anos: "))
meses = int(input("Digite sua idade em Meses: "))
dias = float(input("Digite sua idade em Dias: "))

print("A sua idade em Dias é: ", anos * 365 + meses * 30 + dias)
```

2) Escreva um aplicativo em Python que leia valores positivos aleatórios, que serão
finalizados quando for digitado o valor zero, ou qualquer valor negativo. Como saída de
dados Mostre:

a) O menor Valor

b) O maior Valor

c) A médias dos Valores ímpares

d) A soma dos valores Pares


```python
valor = int(1) 
somapar = float(0)
Npar = int(0)
menor = float(0)
maior = float(0)
somaimpar = float(0)
Nimpar = int(0)

while valor != 0:
  valor = (int(input("Digite um valor: ")))
  if valor < menor:
    menor = valor

  if valor > maior:
    maior = valor

  if valor % 2 == 0:
    somapar += valor
    Npar += 1
    
  else:
    somaimpar += valor
    Nimpar += 1

  if Nimpar > 0:
    media = somaimpar / Nimpar

print("O Menor valor:", menor)
print("O Maior valor:", maior)
print("A Média dos valores ímpares:", media)
print("A Soma dos valores pares:", somapar)
```
