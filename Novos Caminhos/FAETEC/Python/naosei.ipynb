```python
#? ? ? ? ? ? ? ?

print("Insira a sua operação: ")
op = input()
n1 = 0
n2 = 0
if op == "+" :
 n1 = int(input())
 n2 = int(input())
 print("Soma: ", n1+n2)

elif op == "-" :
 n1 = int(input())
 n2 = int(input())
 print("Soma: ", n1-n2)
elif op == "*" :
 n1 = int(input())
 n2 = int(input())
 print("Soma: ", n1*n2)
elif op == "/" or op == "÷" :
 n1 = int(input())
 n2 = int(input())
 print("Soma: ", n1/n2)
elif op == "^" :
 n1 = int(input())
 n2 = int(input())
 print("Soma: ", n1^n2)
else :
  print("Operação nao incluida.")
```
