2- Escreva um aplicativo em python para ler o numero total de eleitores de um municipio, o numero de votos brancos, nulos e validos. calcular e escrever o percentual que cada um representa em relação ao total de eleitores

```python
tot = int(input("Total de Votos: "))
bran = int(input("Total de votos Brancos: "))
nul = int(input("Total de votos Nulos: "))
val = int(input("Total de votos Validos: "))
sm = (bran + nul + val) - tot

if (sm >= 0) :
  print("Total de Votos: ",tot," (Representa 100%)")
  print("Total de votos Validos: ", val, " equivale a ", val / tot * 100,"%")
  print("Total de votos Nulos: ", nul, " equivale a ", nul / tot * 100,"%")
  print("Total de votos Brancos: ", bran, " equivale a ", bran / tot * 100,"%")
else :
  print("Os valores inseridos são invalidos.")

```

3- Faça um programa que leia três valores para a construção de uma tabuada de soma, são eles:

um para ser o valor inicial da tabuada

outro para ser o final da tabuada

o terceiro para ser o valor a ser somado em cada posição da tabuada


```python
val1 = int(input("Digite o valor de Inicio da tabuada: "))
val2 = int(input("Digite o valor Final da tabuada: "))
somaa= int(input("Digite o valor de soma para a tabuada: "))
res = int(0)

while val1 <= val2 :
    res = val1 + somaa
    print(val1, "+", somaa, "=", res)
    val1 += 1
```
