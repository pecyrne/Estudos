```python
i = 1
valores = []
while i <= 10:
    valor = float(input("Digite o valor: "))
    valores += [valor]
    i += 1
maior_valor = max(valores)
menor_valor = min(valores)

soma = maior_valor + menor_valor

print("A soma do maior e do menor valor é:",soma)
```

```python
numerospares = 0
numerosimpares = 0

for i in range(20):
    numero = int(input("Digite o número: "))
    if numero % 2 == 0:
        numerospares += 1
    else:
        numerosimpares += 1
        
print("Quantidade de números pares:",numerospares)
print("Quantidade de números ímpares:",numerosimpares)
```
